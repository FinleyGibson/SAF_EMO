```python
#! /bin/env python
import numpy as np
import wfg
import matplotlib.pyplot as plt

## establish the objective function
M = 3                                   # Number of objectives
kfactor = 2
lfactor = 2
k = kfactor*(M-1)   # position related params
l = lfactor*2           # distance related params

n_dim = k+l
n_obj = M
func = wfg.WFG5

dmv0 = np.array([[1., 1., 2.5]])
dmv1 = np.array([[1., 3.5, 1.5]])
dmv0 = dmv0/np.linalg.norm(dmv0)
dmv1 = dmv1/np.linalg.norm(dmv1)

dmvs = [dmv0, dmv1]

limits = np.zeros((2, n_dim))
limits[1] = np.array(range(1,n_dim+1))*2

def objective_function(x):
    if x.ndim<2:
        x = x.reshape(1,-1)
    return np.array([func(xi, k, M) for xi in x])


N = 500
y = np.zeros((N, M))
for n in range(N):
    z = wfg.random_soln(k, l, func.__name__)
    y[n,:] = func(z, k, M)
```

```python
fig  = plt.figure(figsize=[8, 8])
ax = fig.gca(projection='3d')

ax.scatter(*y.T)
```

```python
import sys
import rootpath
sys.path.append(rootpath.detect())

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pickle

from testsuite.analysis import load_all, plot_measure
from testsuite.surrogates import GP, MultiSurrogate 
```

```python
## load 2 objective results
path = rootpath.detect()+"/experiments/for_paper/wfg6_20bj_6dim/pkl_data/results.pkl"
with open(path, "rb") as file:
    results_2obj = pickle.load(file)[2:4]
```

```python
## load 3 objective results
path = rootpath.detect()+"/experiments/for_paper/wfg6_3obj_6dim/pkl_data/results.pkl"
with open(path, "rb") as file:
    results_3obj = pickle.load(file)[2:4]
```

```python
hv_fig =plt.figure()
hv_ax = hv_fig.gca()

colors= ["C0", "C3"]
for color, result in zip(colors, results_3obj):
    plot_measure(result, "hpv", axis=hv_ax, label=result["name"], color=color, plot_individuals=True)
hv_fig.legend()
hv_fig.suptitle("3 objectives")
```

```python
surrogate_mu = MultiSurrogate(GP, scaled=True)
surrogate_ei = MultiSurrogate(GP, scaled=True)

diffs_2obj_mu = []
means_2obj_mu = []
vars_2obj_mu = []
reals_2obj_mu = []

diffs_2obj_ei = []
means_2obj_ei = []
vars_2obj_ei = []
reals_2obj_ei = []

colors = cm.tab10(np.linspace(0,1,10))
for j, i in enumerate(range(10,100)):
#     color = colors[j]
    xtr_mu, xte_mu = results_2obj[1]["x"][0][:i], results_2obj[1]["x"][0][i:i+1]
    ytr_mu, yte_mu = results_2obj[1]["y"][0][:i], results_2obj[1]["y"][0][i:i+1]
    surrogate_mu.update(xtr_mu, ytr_mu)
    mu_p, var_p = surrogate_mu.predict(xte_mu[0])
    
    means_2obj_mu.append(mu_p)
    vars_2obj_mu.append(var_p)
    reals_2obj_mu.append(yte_mu)
    diffs_2obj_mu.append(yte_mu-mu_p)
    
    xtr_ei, xte_ei = results_2obj[0]["x"][0][:i], results_2obj[0]["x"][0][i:i+1]
    ytr_ei, yte_ei = results_2obj[0]["y"][0][:i], results_2obj[0]["y"][0][i:i+1]
    surrogate_ei.update(xtr_ei, ytr_ei)
    mu_p, var_p = surrogate_ei.predict(xte_ei[0])
    
    means_2obj_ei.append(mu_p)
    vars_2obj_ei.append(var_p)
    reals_2obj_ei.append(yte_ei)
    diffs_2obj_ei.append(yte_ei-mu_p)
    
#     points = ((np.random.randn(1000*2).reshape(1000,2)*std_p)+mu_p)
    
    
#     ax_scat.scatter(points[:,0], points[:,1], c=color, alpha=0.1, s=10)
#     ax_scat.scatter(mu_p[:,0], mu_p[:,1], s=70, c=color, marker="*")
#     ax_scat.scatter(yte[0,0], yte[0,1], s=70, c=color, marker = "x")
#     ax_scat.set_xlim([0,5])
#     ax_scat.set_ylim([0,10])

```

```python
diffs_3obj_mu = np.array(diffs_3obj_mu).squeeze(1)
diffs_3obj_ei = np.array(diffs_3obj_ei).squeeze(1)
```

```python
obj1_optimism_mu = diffs_2obj_mu[:,0]<0.
obj2_optimism_mu = diffs_2obj_mu[:,1]<0.
obj1_optimism_ei = diffs_2obj_ei[:,0]<0.
obj2_optimism_ei = diffs_2obj_ei[:,1]<0.
```

```python
q1_mu = np.sum([i and not j for i, j in zip(obj1_optimism_mu, obj2_optimism_mu)])/len(obj1_optimism_mu)
q2_mu = np.sum([not i and not j for i, j in zip(obj1_optimism_mu, obj2_optimism_mu)])/len(obj1_optimism_mu)
q3_mu = np.sum([i and j for i, j in zip(obj1_optimism_mu, obj2_optimism_mu)])/len(obj1_optimism_mu)
q4_mu = np.sum([not i and j for i, j in zip(obj1_optimism_mu, obj2_optimism_mu)])/len(obj1_optimism_mu)
```

```python
q1_ei = np.sum([i and not j for i, j in zip(obj1_optimism_ei, obj2_optimism_ei)])/len(obj1_optimism_ei)
q2_ei = np.sum([not i and not j for i, j in zip(obj1_optimism_ei, obj2_optimism_ei)])/len(obj1_optimism_ei)
q3_ei = np.sum([i and j for i, j in zip(obj1_optimism_ei, obj2_optimism_ei)])/len(obj1_optimism_ei)
q4_ei = np.sum([not i and j for i, j in zip(obj1_optimism_ei, obj2_optimism_ei)])/len(obj1_optimism_ei)
```

```python
assert np.sum([q1_mu, q2_mu, q3_mu, q4_mu]) == 1.0
assert np.sum([q1_ei, q2_ei, q3_ei, q4_ei]) == 1.0
```

```python
fontdict = {"fontsize": 50, "weight": "bold", "ha": "center", "va": "center"}

fig_2d = plt.figure(figsize=[10, 10])
ax_2d = fig_2d.gca()
ax_2d.scatter(*diffs_2obj_mu.T, c="C0", s=40, label="$\mu$")
ax_2d.scatter(*diffs_2obj_ei.T, c="C1", s=40, label="ei")
# plt.scatter(diffs_3obj[:,0], diffs_3obj[:,1],c="C1")
ax_2d.text(ax_2d.get_xlim()[0]/2, ax_2d.get_ylim()[1]/2+0.25, "{:.2f}%".format(q1_mu*100), alpha=0.3, color="C0", fontdict=fontdict)
ax_2d.text(ax_2d.get_xlim()[1]/2, ax_2d.get_ylim()[1]/2+0.25, "{:.2f}%".format(q2_mu*100), alpha=0.3, color="C0", fontdict=fontdict)
ax_2d.text(ax_2d.get_xlim()[0]/2, ax_2d.get_ylim()[0]/2+0.25, "{:.2f}%".format(q3_mu*100), alpha=0.3, color="C0", fontdict=fontdict)
ax_2d.text(ax_2d.get_xlim()[1]/2, ax_2d.get_ylim()[0]/2+0.25, "{:.2f}%".format(q4_mu*100), alpha=0.3, color="C0", fontdict=fontdict)
ax_2d.text(ax_2d.get_xlim()[0]/2, ax_2d.get_ylim()[1]/2-0.25, "{:.2f}%".format(q1_ei*100), alpha=0.3, color="C1", fontdict=fontdict)
ax_2d.text(ax_2d.get_xlim()[1]/2, ax_2d.get_ylim()[1]/2-0.25, "{:.2f}%".format(q2_ei*100), alpha=0.3, color="C1", fontdict=fontdict)
ax_2d.text(ax_2d.get_xlim()[0]/2, ax_2d.get_ylim()[0]/2-0.25, "{:.2f}%".format(q3_ei*100), alpha=0.3, color="C1", fontdict=fontdict)
ax_2d.text(ax_2d.get_xlim()[1]/2, ax_2d.get_ylim()[0]/2-0.25, "{:.2f}%".format(q4_ei*100), alpha=0.3, color="C1", fontdict=fontdict)
ax_2d.axhline(0., c="k")
ax_2d.axvline(0., c="k")
fig_2d.legend()
```

```python
a = 2
b = 0.6
c = 1
r = 1
```

```python
X = randn(2000, 3)
X = X*asarray([a, b, c])
fig = figure()
ax = fig.gca(projection='3d')
ax.scatter(*X.T, marker='.')
```

```python
%matplotlib qt
```

```python
Y = []
for x in X:
    lmbda = r**2/(x[0]**2/a**2 + x[1]**2/b**2 + x[2]**2/c**2)
    lmbda = sqrt(lmbda)
    y = x*lmbda
    Y.append(np.abs(y))
Y = asarray(Y)
    
fig = figure()
ax = fig.gca(projection='3d')
ax.set_box_aspect([a,b,c]) 
ax.scatter(*Y.T, marker='.')

```

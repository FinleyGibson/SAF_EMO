```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_points = 1000
a, b = 5, 1
x = np.random.multivariate_normal([0, 0], np.diag([a**2., b**2]), size = n_points)
angs = np.linspace(0, 2*np.pi, 100)
ox, oy = a*np.sin(angs), b*np.cos(angs)
```

```python
fig = plt.figure(figsize=[20, 20])
ax = fig.gca()
ax.scatter(*x.T, alpha=0.3, s=5)
ax.set_aspect('equal')
ax.plot(ox, oy, c="C1", linewidth=2)
ax.scatter([0], [0], c="C3", s=10)

vt = np.linspace(0, 2*np.pi, 11)[:-1]
ax.scatter(a*np.sin(vt), b*np.cos(vt), s=50, c="C4")
```

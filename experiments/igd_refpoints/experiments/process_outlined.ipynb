```python
y_at = attainment_sample(y, 5000)
```

```python
fig = plt.figure(figsize=[8, 8])
ax = fig.gca(projection="3d")
ax_format(ax, axes=axes, vp=[35, 15])
ax.scatter(*y_at.T, s=2)

save_fig(fig, "process_04_attainment_points_3D_full")
```

```python
aa = distance_matrix(y, y_at)
```

```python
D = aa.min(axis=0)
D2 = aa.min(axis=1)

print(D.shape)
print(D2.shape)
```

```python
print(D.max())
print(D2.max())
```

```python
bb = [sum(D<i) for i in np.linspace(0.01, 0.5, 200)]
bb2 = [sum(D2<i) for i in np.linspace(0.01, 0.5, 200)]
```

```python
fig = plt.figure()
ax = fig.gca()
ax.plot(np.linspace(0.01, 0.5, 200), np.array(bb)/len(D), label="Pareto surface points")
ax.plot(np.linspace(0.01, 0.5, 200), np.array(bb2)/len(D2), label="Attainment surface points")

ax.set_xlabel("Threshhold distance")
ax.set_ylabel("n below threshold/n totaal")

thresh = 0.1

ax.axvline(thresh, linestyle="--", c="r", label=r"~$100\%$ Pareto points within threshold")

ax.legend()
ax.set_xlim([0, 0.5])
ax.set_ylim([0, 1.1])

save_fig(fig, "process_05_threshholding_decision")
```

```python
fig = plt.figure(figsize=[8, 8])
ax = fig.gca(projection="3d")
ax_format(ax, axes=axes, vp=[35, 15])
ax.scatter(*y_at[D<0.1].T, s=2, c="C0", label="Attainment points < nn threshold")
ax.scatter(*y_at[D>0.1].T, s=2, c="C1", label="Attainment points > nn threshold")
ax.legend()

save_fig(fig, "process_06_attainment_thresholded")
```

```python
fig = plt.figure(figsize=[8, 8])
ax = fig.gca(projection="3d")
ax_format(ax, axes=axes, vp=[35, 10])
ax.scatter(*y_at[D<0.1].T, s=2, c="C0", label="Evenly distributed samples close to Pareto front")
ax.legend()

save_fig(fig, "process_07_resulting_points_ang1")
```

```python
import sys
import rootpath
sys.path.append(rootpath.detect())

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import wfg
from testsuite.utilities import Pareto_split
from experiments.igd_refpoints.igd_utilities import *
from scipy.spatial import distance_matrix
```

```python
try:
    with open('data/wfg3_2d_refpoints', 'rb') as infile:
        yf_2 = np.load(infile)
    yf_2_loaded = True
    print("2D data: loaded")
except FileNotFoundError:
    yf_2_loaded = False
    print("2D data: not found")
```

```python
if not yf_2_loaded:
    x = np.linspace(0, 2, 2000)
    y = -2*x+4
    
    yf_2 = np.vstack((x, y)).T
yf_2.shape
```

```python
func = wfg.WFG3
n_obj = 2
l = 3
k = 3
n_dim = k+l

M = n_obj
N = 200 
y = np.zeros((N, n_obj))
x = np.zeros((N, n_dim))
for n in range(N):
    z = wfg.random_soln(k, l, func.__name__)
    y[n,:] = func(z, k, M)
    x[n,:] = z

a, b = np.round(y.max(axis=0), 1)
    
fig = plt.figure(figsize=[8, 8])
ax0 = fig.add_subplot(121)
ax1 = fig.add_subplot(122)
ax_format(ax0, [a, b])
ax_format(ax1, [a, b])


ax0.plot(*yf_2[np.argsort(yf_2[:,0])].T, c="C0", alpha=0.3,  label="Pareto surface")
ax0.scatter(*y.T, s=20, c="C1", marker="x", label="points drawn from WFG3")

ax1.plot(*yf_2[np.argsort(yf_2[:,0])].T, c="C0", alpha=0.3, label="Pareto surface")
ax1.scatter(*yf_2.T, s=20, c="C1", marker="x", label="computed points")

ax0.legend()
ax1.legend()
```

```python
if yf_2_loaded is False:
    with open('data/wfg3_2d_refpoints', 'wb') as outfile:
        np.save(outfile, yf_2)
```
